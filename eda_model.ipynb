Parameters
- test_size = 0.2
    - 20% of your data  goes to test set

Random_state=42
- using the same random_state will always produce the same split( example : 20% and 80%) every time you run the code 
- it makes your experiments reproducible 
- the number itself doesn't matter. You can use any number (0,7,23,...)
